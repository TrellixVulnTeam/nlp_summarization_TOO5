```
bart = torch.hub.load('pytorch/fairseq', 'bart.large.cnn')
```

```
import pickle

# PATH = '/Users/alexgaskell/Technical-Writer-Assistant/finished_files/'
# PATH = '/Users/alexgaskell/neural-summ-cnndm-pytorch/deepmind/validate_set/'
PATH = '/content/drive/My Drive/Documents/Imperial/Individual_project/'

bart.cuda()
bart.half()
bart.eval()
count = 1
bsz = 32

# Load data
with open(PATH + 'pj1000.pkl', "rb") as f:
    source = pickle.load(f)

# Process into mini batches
batches = []
for j in range(0, len(source), bsz):
    mini_batch = []
    for k in range(32):
        mini_batch.append(source[k][0][1])
    batches.append(mini_batch)

with open(PATH + 'pj1000.hyp_remote', "w") as fout:
    print('Beginning')
    hypotheses_batch = bart.sample(batches[0], beam=4, lenpen=2.0, max_len_b=140, min_len=55, no_repeat_ngram_size=3)
    print('Finished processing')
    print(len(hypotheses_batch))

    for hypothesis in hypotheses_batch:
        fout.write(hypothesis + '\n')
        fout.flush()
```
